```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
out_path = "train_v2/labels_png/"
labels_path = "train_v2/labels/"
labels = [file for file in os.listdir(labels_path) if file.endswith(".npy")]
```

```python
len(labels)
```

```python
for file in labels:
    arr = np.load(os.path.join(labels_path, file))
    arr = arr.astype(np.uint8)
    image = Image.fromarray(arr, mode='L')
    png_filename = os.path.splitext(file)[0] + '.png'
    png_filepath = os.path.join(out_path, png_filename)
    image.save(png_filepath)
    # plt.imshow(image)  # cmap='gray' for grayscale display
    # plt.show()
    
```

```python
for file in labels:
    png_filename = os.path.splitext(file)[0] + '.png'
    img = Image.open(out_path + png_filename)
    plt.imshow(img)  # cmap='gray' for grayscale display
    plt.show()
```
